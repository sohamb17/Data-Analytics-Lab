**Hypothesis testing**

This hypothesis test will determine whether the mean of runs scored in the first innings of a match is equal to 150 or greater than it.

So, Null hypothesis: u = 150

and Alternate hypothesis: u > 150

```python
df['first_innings'].describe()
```

Sample of 8 matches

```python
sample_df = df['first_innings'].sample(8)
sample_df
```

```python
std_dev = df['first_innings'].std()
print(f's = {std_dev}')
print(f'Sample mean = {sample_df.mean()}')
n = 8
print(f'n = {n}')
std_err = std_dev / (n ** 0.5)
print(f'Standard error = {std_err}')
t, p = sp.stats.ttest_1samp(sample_df, 150)
print(f'p-value = {p}')
print(f't-statistic = {t}')
if t > 1.895:
  print('Null hypothesis rejected')
else:
  print('Null hypothesis accepted')
if p > 0.1:
  print("Non significant evidence against Null hypothesis")
elif 0.1 >= p > 0.05:
  print("Marginally significant evidence against Null hypothesis") 
elif 0.05 >= p > 0.01:
  print("Significant evidence against Null hypothesis") 
else:
  print("Highly significant evidence against Null hypothesis")
```

The t-test outputs that the null hypothesis was accepted, meaning that the mean of runs scored in the first innings of a match is indeed 150 runs.

**Example for hypothesis testing**

Null hypothesis: u = 30

Alternate hypothesis: u != 30

Mean age is different from 30 if null hypothesis is rejected.

n = 10

mean (x_bar) = 27

var = 20

alpha = 0.05

z = (x_bar - 30) / (var / n) ** 0.5

z = -2.12

Null hypothesis is rejected if z > z(1-alpha/2)

i.e., if z < -z(1 - 0.025)

i.e., if z < -z(0.975)

z(0.975) = 1.96

Since, -2.12 < -1.96, the null hypothesis can be rejected.

Therefore, we accept the alternate hypothesis, i.e. u is not equal to 30.

Given p_value = 0.0340

Since, p_value < alpha, the null hypothesis can be rejected.

<br/><br/>

Null hypothesis: u = 30

Alternate hypothesis: u < 30

z = (x_bar - 30) / (var / n) ** 0.5

z = -2.12

Null hypothesis is rejected if z > z(1-alpha)

i.e., if z < -z(1 - 0.05)

i.e., if z < -z(0.95)

z(0.975) = 1.645

Since, -2.12 < -1.645, the null hypothesis can be rejected.

Therefore, we accept the alternate hypothesis, i.e. u is smaller than 30.
