```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
with open('/content/drive/MyDrive/groceries.csv') as G:
    groceries = G.read()

groceries = [i.split(',') for i in groceries.split('\n')]
groceries_list = [j for i in groceries for j in i]
groceries_list
```

```python
df = pd.DataFrame(pd.Series(groceries), columns = ['groceries'])
df
```

```python
X = df['groceries'].str.join(sep='*').str.get_dummies(sep='*')
X
```

```python
frequent_itemsets = apriori(X, min_support=0.001, max_len=4, use_colnames=True)
frequent_itemsets.head()
```

min support = 0.001 = 0.1%

```python
asso_rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.9)
asso_rules.columns
```

min confidence = 0.9 = 90%

```python
asso_rules.head()
```
